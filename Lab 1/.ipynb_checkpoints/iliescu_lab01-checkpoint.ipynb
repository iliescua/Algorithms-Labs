#### Introduction:
In this lab we were tasked with writing an insertion sort algorithm and then testing several cases against it to see the time complexity of the algorithm. We needed to test against varying sized lists consisting of a random list, a list that had already been sorted from smallest to largest, and a list that was sorted in the reverse order. The algorithm was tasked with sorting each list ten times for a specific size before moving on to the next size and repeating the process. The system clock was used in order to determine the duration of each sort, and then a graph was constructed to display the runtime vs the log10(list size). The results showed which list was the fastest to sort and which was the slowest to sort by creating a trend amongst the varying sizes. The fastest list to sort was the one that was already arranged from largest to smallest, the list that took a reasonable amount of time was the randomly shuffled list, and the slowest by far was the reverse sorted list.

#### Questions:
1. The run times between the three cases varied quite drastically. The pattern between the three was that the ordered sort was consistently the fastest regardless of the input size. The reversed sort was always incredibly slow especially as the list size increased, and the random shuffle seemed to be completed twice as fast as the reversed sort, so it was the average case. 
2. The inputs caused the run time to vary because when entering the sorted loop, the program only needed to run through the data set once which means that its run time varied only with the size of the loop. When entering the reverse sort, the run time was at its worst because it had to iterate through the entire data set twice essentially, since it had to constantly go all the way back to the start every time it moved one value over. The random shuffle was a mix of both cases, hence why it presented the average run time result. 
3. The big-Oh run time of the insertion sort is: best case O(n), average case O(n^2), worst case O(n^2). These run time results are consistent with my benchmarks because in the best case (sorted list) the run time was only dependent on the size of the list because it still needed to iterate through the entire list, which is an O(n) operation. For the worst case (reverse sorted) the run time is dependent on running through the entire list twice because it needs to check every value and each time it checks a new value it needs to run back through the list to place it, meaning that it would be O(n2). The average run time (random shuffle) makes sense because while it can be better than O(n2) if the shuffle ends up being more sorted, it will more than likely end up being O(n2), especially with sufficiently long lists.

```python
from sorting import insertion_sort
import matplotlib.pyplot as plt
import numpy as np
import random as rand
import time
```

```python
def benchmark(lst):
    times = [] #Used to store the duration time
    n_trials = 10 #Number of trials for each list
    for i in range(n_trials):
        copy = lst[:] #Duplicate list was made so that the order isn't altered
        start = time.clock()
        insertion_sort(copy)
        end = time.clock()
        elapsed = end - start
        times.append(elapsed)
    avg = np.mean(times) #Calculates the average run time
    print("Average Time: %.10f seconds\n" %(avg))
    return avg
```

```python
def plotMaker(trials, result_shuffle, result_sort, result_reverse):
    #This method is called to simply create a plot of the results
    plt.plot(np.log10(trials), result_shuffle, label="Shuffle List")
    plt.plot(np.log10(trials), result_sort, label="Sort List")
    plt.plot(np.log10(trials), result_reverse, label="Reverse List")
    plt.xlabel("List Size (log10)", fontsize=12)
    plt.ylabel("Running Time (s)", fontsize=12)
    plt.title("Insertion Sort: List Size vs. Run Time", fontsize=14)
    plt.legend()
```

```python
trials = [100, 1000, 10000, 100000] #List Sizes
lst = [] 
result_shuffle = [] 
result_sort = []
result_reverse = []

for length in trials: #Iterates so long as there are more lists to do
    lst = [rand.random() for i in range(length)] #Creates a list of random values
    
    #Shuffled List
    rand.shuffle(lst) 
    print("\nRandom Shuffle for Size: %d" %(length))
    result_shuffle.append(benchmark(lst)) #Calls the benchmark and stores average time 
    
    #Sorted List Smallest->Largest
    lst.sort()
    print("\nSort for Size: %d" %(length))
    result_sort.append(benchmark(lst)) #Calls the benchmark and stores average time
    
    #Reverse Order List Largest->Smallest
    lst.reverse()
    print("\nReverse Sort for Size: %d" %(length))
    result_reverse.append(benchmark(lst)) #Calls the benchmark and stores average time
plotMaker(trials, result_shuffle, result_sort, result_reverse)
```
