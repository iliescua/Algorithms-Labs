#### Introduction:
In this lab we were tasked with writing a merge sort algorithm and then testing several cases against it to see the time complexity of the algorithm. We needed to test against varying sized lists consisting of a random list, a list that had already been sorted from smallest to largest, and a list that was sorted in the reverse order. The algorithm was tasked with sorting each list ten times for a specific size before moving on to the next size and repeating the process. The system clock was used in order to determine the duration of each sort, and then a graph was constructed to display the runtime vs the log10(list size). The results showed that the merge sort ran almost identically for each list type meaning that the way in which the lists were sorted prior to applying the algorithm made no impact on the time complexity and the run was only impacted by the list's size. Then the merge sort was compared to the insertion sort in which the merge sort algorithm ran significantly faster in almost every instance because merge sort runs in O(nlogn) no matter what, whereas insertion sort runs in O(n^2) except for the case where the list being entered is already sorted from smallest to largest, because then insertion sort runs in O(n).

```python
import sorting 
import matplotlib.pyplot as plt
import numpy as np
import random as rand
import time
```

```python
def benchmark_merge(lst):
    times = [] #Used to store the duration time
    n_trials = 10 #Number of trials for each list
    for i in range(n_trials):
        copy = lst[:] #Duplicate list was made so that the order isn't altered
        start = time.clock()
        sorting.merge_sort(copy, 0, len(lst) - 1)
        end = time.clock()
        elapsed = end - start
        times.append(elapsed)
    avg = np.mean(times) #Calculates the average run time
    print("Merge Sort Average Time: %.10f seconds" %(avg))
    return avg
```

```python
def benchmark_insertion(lst):
    times = [] #Used to store the duration time
    n_trials = 10 #Number of trials for each list
    for i in range(n_trials):
        copy = lst[:] #Duplicate list was made so that the order isn't altered
        start = time.clock()
        sorting.insertion_sort(copy)
        end = time.clock()
        elapsed = end - start
        times.append(elapsed)
    avg = np.mean(times) #Calculates the average run time
    print("Insertion Sort Average Time: %.10f seconds\n" %(avg))
    return avg
```

```python
def plotMerge(trials, result_shuffle, result_sort, result_reverse):
    #This method is called to simply create a plot of the results
    plt.plot(np.log10(trials), result_shuffle, label="Shuffled List")
    plt.plot(np.log10(trials), result_sort, label="Sorted List")
    plt.plot(np.log10(trials), result_reverse, label="Reversed List")
    plt.xlabel("List Size (log10)", fontsize=12)
    plt.ylabel("Running Time (s)", fontsize=12)
    plt.title("Merge Sort: List Size vs. Run Time", fontsize=14)
    plt.legend()
```

```python
trials = [100, 1000, 10000, 100000] #List Sizes
lst = [] 
result_shuffle = [] 
result_sort = []
result_reverse = []

for length in trials: #Iterates so long as there are more lists to do
    lst = [rand.random() for i in range(length)] #Creates a list of random values
    
    #Shuffled List
    rand.shuffle(lst) 
    print("\nRandom Shuffle for Size: %d" %(length))
    result_shuffle.append(benchmark_merge(lst)) #Calls the benchmark and stores average time 
    
    #Sorted List Smallest->Largest
    lst.sort()
    print("\nSort for Size: %d" %(length))
    result_sort.append(benchmark_merge(lst)) #Calls the benchmark and stores average time
    
    #Reverse Order List Largest->Smallest
    lst.reverse()
    print("\nReverse Sort for Size: %d" %(length))
    result_reverse.append(benchmark_merge(lst)) #Calls the benchmark and stores average time
plotMerge(trials, result_shuffle, result_sort, result_reverse)
```

```python
def plotMaker(trials, resultM_shuffle, resultI_shuffle, resultM_sorted, resultI_sorted, resultM_reverse, resultI_reverse):
    #This method is called to simply create a plot of the results

    #Shuffled List Plot
    plt.plot(np.log10(trials), resultM_shuffle, label="Merge Sort")
    plt.plot(np.log10(trials), resultI_shuffle, label="Insertion Sort")
    plt.xlabel("List Size (log10)", fontsize=12)
    plt.ylabel("Running Time (s)", fontsize=12)
    plt.title("Shuffled List: Merge Sort vs. Insertion Sort", fontsize=14)   
    plt.legend()
    plt.show()

    #Sorted List Plot
    plt.plot(np.log10(trials), resultM_sorted, label="Merge Sort")
    plt.plot(np.log10(trials), resultI_sorted, label="Insertion Sort")
    plt.xlabel("List Size (log10)", fontsize=12)
    plt.ylabel("Running Time (s)", fontsize=12)
    plt.title("Sorted List: Merge Sort vs. Insertion Sort", fontsize=14)   
    plt.legend()
    plt.show()

    #Reverse Sorted List Plot
    plt.plot(np.log10(trials), resultM_reverse, label="Merge Sort")
    plt.plot(np.log10(trials), resultI_reverse, label="Insertion Sort")
    plt.xlabel("List Size (log10)", fontsize=12)
    plt.ylabel("Running Time (s)", fontsize=12)
    plt.title("Reverse Sorted List: Merge Sort vs. Insertion Sort", fontsize=14)   
    plt.legend()
    plt.show()
```

```python
trials = [100, 1000, 10000, 100000] #List Sizes
lst = [] 
#Stores the avg runtimes for the diff lists
resultM_shuffle = [] 
resultM_sorted = []
resultM_reverse = []
resultI_shuffle = [] 
resultI_sorted = []
resultI_reverse = []


for length in trials: #Iterates so long as there are more lists to do
    lst = [rand.random() for i in range(length)] #Creates a list of random values
    
    #Shuffled List
    rand.shuffle(lst) 
    print("\nRandom Shuffle for Size: %d" %(length))
    resultM_shuffle.append(benchmark_merge(lst)) 
    resultI_shuffle.append(benchmark_insertion(lst))
    
    #Sorted List Smallest->Largest
    lst.sort()
    print("\nSort for Size: %d" %(length))
    resultM_sorted.append(benchmark_merge(lst)) 
    resultI_sorted.append(benchmark_insertion(lst))
    
    #Reverse Order List Largest->Smallest
    lst.reverse()
    print("\nReverse Sort for Size: %d" %(length))
    resultM_reverse.append(benchmark_merge(lst)) 
    resultI_reverse.append(benchmark_insertion(lst))

plotMaker(trials, resultM_shuffle, resultI_shuffle, resultM_sorted, resultI_sorted, resultM_reverse, resultI_reverse)
```
